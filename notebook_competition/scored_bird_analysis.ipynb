## idea to work on co occurance

- make a Zero square matrics  with no of birds X no of birds
- add 1 to all the pair 
- add 1 to self pair (b1, b1)




```python
f = open('../data/birdclef-2022/bird_to_idx.json')
bird_to_idx = json.load(f)
```

```python
def matrix_creator(df, bb_index):
    co_occurance_mat = np.zeros(shape = (len(bb_index), len(bb_index)))
    pairs = []
    
    for index, row in df.iterrows():
        prim = bb_index[row['primary_label']]
        pairs.append([prim, prim])
        if len(row['secondary_labels']) > 0:
            for sec in row['secondary_labels']:
                try: # sometime the socred birds occure with other birds
                    pairs.append([prim, bb_index[sec]])
                except:
                    pass
    for i in pairs:
        co_occurance_mat[i[0]][i[1]] += 1
        if i[0] != i[1]:
            co_occurance_mat[i[1]][i[0]] += 1
    return co_occurance_mat
        
```

## co occurance for the entire training dataset

```python
co_occurance_mat = matrix_creator(train_df, bird_to_idx)
co_occurance_mat = co_occurance_mat/co_occurance_mat.max(axis=1)[:,None]
plt.figure(figsize=(25, 25))
sns.heatmap(co_occurance_mat,xticklabels=range(len(bird_to_idx)),yticklabels=range(len(bird_to_idx)))
```
